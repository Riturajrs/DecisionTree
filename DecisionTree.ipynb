```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas.core.frame import DataFrame
from  sklearn.preprocessing import LabelEncoder


df = pd.read_csv('balance-scale.csv')
df['Class'] = LabelEncoder().fit_transform(df['Class'])
X = df.drop(['Class'],axis=1)
y = df['Class']
# print(X)
# print(y)
from sklearn.model_selection import train_test_split
from sklearn.tree import  DecisionTreeClassifier
acc = 0
acc_list = []
i_list = []
for i in range(10,90):
  X_train, X_test, y_train, y_test = train_test_split(X,y,train_size=(i/100))
  model = DecisionTreeClassifier(criterion="entropy")
  model.fit(X_train, y_train)
  i_list.append(i)
  acc_list.append(model.score(X_test,y_test))
  acc = max(acc,model.score(X_test,y_test))  
plt.xlabel('Train size in %')
plt.ylabel('Accuracy')
plt.plot(i_list,acc_list)
plt.show()
print("Model accuracy: ",acc*100)
```
